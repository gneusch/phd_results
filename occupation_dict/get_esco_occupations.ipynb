```python
class EscoOccupation():
    def __init__(self, class_name, p_label, a_labels, uri, description):
        self.p_label = p_label
        self.a_labels = a_labels        
        self.uri = uri        
        self.description = description        
```

```python
def getOccupationsFromNode(node):    
    a_labels = []
    try:
        a_labels = node.properties()['alternativeLabel']['en']
    except KeyError:
        a_labels = None
        
    result = [
        EscoOccupation(
            node.properties()['className'], 
            node.properties()['preferredLabel']['en'], 
            a_labels, 
            node.link(rel='self')['href'],
            node.properties()['description']['en']['literal']
        )
    ]
    
    links = node.links(rel='narrowerConcept') + node.links(rel='narrowerOccupation')
    
    if not links:
        return result
    else:        
        for l in links:
            try:
                result = result + getOccupationsFromNode(l.follow())
            except AttributeError:
                print(l)
                print(type(l))
                raise AttributeError
            except TypeError: 
                print(l)
                print(type(l))
                raise TypeError
        return result  
```

```python
technicians = getOccupationsFromNode(HALEasy('https://ec.europa.eu/esco/api/resource/concept?uri=http://data.europa.eu/esco/isco/C35&language=en'))  
professionals = getOccupationsFromNode(HALEasy('https://ec.europa.eu/esco/api/resource/concept?uri=http://data.europa.eu/esco/isco/C25&language=en'))
mid_managers = getOccupationsFromNode(HALEasy('https://ec.europa.eu/esco/api/resource/concept?uri=http://data.europa.eu/esco/isco/C133&language=en'))
rnd_managers = getOccupationsFromNode(HALEasy('https://ec.europa.eu/esco/api/resource/concept?uri=http://data.europa.eu/esco/isco/C1223&language=en'))
sci_professionals = getOccupationsFromNode(HALEasy('https://ec.europa.eu/esco/api/resource/concept?uri=http://data.europa.eu/esco/isco/C215&language=en'))
app_eng = getOccupationsFromNode(HALEasy('https://ec.europa.eu/esco/api/resource/concept?uri=http://data.europa.eu/esco/occupation/fde4372d-8293-42de-afbb-8b5a9b274fe7&language=en'))
stat_technicians = getOccupationsFromNode(HALEasy('https://ec.europa.eu/esco/api/resource/concept?uri=http://data.europa.eu/esco/isco/C3314&language=en'))
statistican = getOccupationsFromNode(HALEasy('https://ec.europa.eu/esco/api/resource/concept?uri=http://data.europa.eu/esco/occupation/ac8b3cd1-a127-4e6a-8208-5cfcf7111955&language=en'))
office_supervisionrs = getOccupationsFromNode(HALEasy('https://ec.europa.eu/esco/api/resource/concept?uri=http://data.europa.eu/esco/isco/C3341&language=en'))
```

```python
esco_occupation = technicians + professionals + mid_managers + rnd_managers + sci_professionals + app_eng + stat_technicians + statistican + office_supervisionrs
print(len(esco_occupation))
```

```python
for o in esco_occupation:
    o.uri = o.uri[o.uri.find("=") + 1: o.uri.find("&")]
```

```python
with open('esco_occupations.json', 'w') as outfile:
    outfile.write("%s\n" % json.dumps([o.__dict__ for o in esco_occupation]))
```

```python
occupation_labels = []
for oc in esco_occupation:
    occupation_labels.append(oc.p_label)
    if oc.a_labels:
        occupation_labels = occupation_labels + oc.a_labels

print(len(occupation_labels))
print(occupation_labels)
with open('esco_occupations.txt', 'w') as outfile:
    for o in occupation_labels:
        outfile.write("%s\n" % o)
```
